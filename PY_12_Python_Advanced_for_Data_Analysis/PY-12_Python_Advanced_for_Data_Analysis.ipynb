```python
exclude_columns = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

category_columns = []

for column in melb_df.columns:
    if column in exclude_columns:
        continue
    
    unique_count = melb_df[column].nunique()
    print(f"{column} : {unique_count}")
    
    if unique_count < 150:
        melb_df[column] = melb_df[column].astype('category')
        category_columns.append(column)
```

```python
# Преобразование столбца даты в datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

# Фильтрация по дате с 1 мая по 1 сентября 2017 года включительно
start_date = pd.to_datetime('2017-05-01')
end_date = pd.to_datetime('2017-09-01')
filtered_df = melb_df[(melb_df['Date'] >= start_date) & (melb_df['Date'] <= end_date)]

# Группировка по компании SellerG и расчет суммы Price
grouped = filtered_df.groupby('SellerG')['Price'].sum().reset_index()

# Сортировка по возрастанию суммы продаж
sorted_grouped = grouped.sort_values(by='Price')

# Получение компании с наименьшей выручкой
min_seller = sorted_grouped.iloc[0]
print(f"Компания с наименьшей выручкой: {min_seller['SellerG']}, сумма: {min_seller['Price']}")
```
