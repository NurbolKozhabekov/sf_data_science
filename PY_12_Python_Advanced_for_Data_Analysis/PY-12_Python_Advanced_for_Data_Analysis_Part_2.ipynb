```python
import pandas as pd
import os
import re

def concat_user_files(path):
    """
    Объединяет информацию из всех CSV-файлов в указанной директории.
    
    Args:
        path (str): Путь до директории с CSV-файлами
        
    Returns:
        pd.DataFrame: Объединенная таблица с уникальными пользователями,
                     отсортированная по числовой части user_id
    """
    # Проверяем существование директории
    if not os.path.exists(path):
        raise FileNotFoundError(f"Директория {path} не найдена")
    
    # Получаем список всех CSV-файлов в директории
    csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]
    
    if not csv_files:
        raise ValueError(f"В директории {path} нет CSV-файлов")
    
    # Список для хранения DataFrame'ов из каждого файла
    dataframes = []
    
    # Читаем каждый CSV-файл
    for file in csv_files:
        file_path = os.path.join(path, file)
        try:
            df = pd.read_csv(file_path)
            # Проверяем наличие необходимых колонок
            if 'user_id' in df.columns and 'image_url' in df.columns:
                dataframes.append(df[['user_id', 'image_url']])
            else:
                print(f"Предупреждение: файл {file} не содержит колонки 'user_id' или 'image_url'")
        except Exception as e:
            print(f"Ошибка при чтении файла {file}: {e}")
    
    if not dataframes:
        raise ValueError("Не найдено ни одного корректного CSV-файла с нужными колонками")
    
    # Объединяем все DataFrame'ы
    combined_df = pd.concat(dataframes, ignore_index=True)
    
    # Удаляем дубликаты
    combined_df = combined_df.drop_duplicates()
    
    # Обновляем индексы
    combined_df = combined_df.reset_index(drop=True)
    
    # Функция для извлечения числовой части из user_id
    def extract_numeric_part(user_id):
        """Извлекает числовую часть из user_id для сортировки"""
        # Ищем все числа в строке и берем первое найденное
        numbers = re.findall(r'\d+', str(user_id))
        return int(numbers[0]) if numbers else 0
    
    # Создаем временную колонку для сортировки
    combined_df['sort_key'] = combined_df['user_id'].apply(extract_numeric_part)
    
    # Сортируем по числовой части user_id
    combined_df = combined_df.sort_values('sort_key')
    
    # Удаляем временную колонку и обновляем индексы
    combined_df = combined_df.drop('sort_key', axis=1).reset_index(drop=True)
    
    return combined_df
```

```python
import pandas as pd
import os

def concat_user_files(path):
    data = pd.DataFrame()
    file_names = os.listdir(path)
    file_names.sort()
    for file in file_names:
        tmp_data = pd.read_csv(path + '/' + file)
        data = pd.concat([data, tmp_data], axis=0, ignore_index=True)
    data = data.drop_duplicates()
    return data
```
